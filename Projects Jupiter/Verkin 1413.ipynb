## 11.1 Задание
Из заданного числа вычли сумму его цифр. Из результата вновь вычли сумму его цифр и так далее. Через сколько таких действий получится нуль?
## 11.2 Блок-схема
## 11.3 Листинг программы

````python
def minussum(number):
    numberstr=str(number) # конвертация в строку числа
    sumnum = len(numberstr) # количество цифр в числе
    ressum = 0
    for i in range(0, sumnum):
        ressum=ressum+int(numberstr[i])            
    number=int(number)-int(ressum)
    return number
    
def howmuchiterations(number):
    iterations=0
    while number!=0:
        number=minussum(number)
        iterations=iterations+1
    return iterations
        

inputed = 0
inputed = input('Введите число ')
print('Количество итераций для нуля: ', howmuchiterations(inputed))
````

## 12.1 Задание (предусмотреть два режима работы: обработка тестового файла и сгенерированного, причем уже в этой программе используем функции)

Сформировать текстовый файл Data.txt из заглавных букв A, B, C, D, E, F латинского алфавита и десятичных цифр. Найти номер строки, в которой находится самое большое нечётное число, и определить в ней самую длинную возрастающую последовательность из букв. Если строк с максимальным нечётным числом несколько, то выбрать ту, которая находится в файле позже.

## 12.2 Листинг программы

````python
import random as rand

def randfill():
    
    # Заполнение файла Data.txt
    
    n = int(input('Введите количество строк '))
    f = open('Data.txt', 'w')
    for _ in range(n):
        cnt = rand.randrange(30, 150+1) # число символов в строке
        s = ''.join([rand.choice([chr(rand.randrange(ord('A'), ord('G')+1)),chr(rand.randrange(ord('0'), ord('9')+1))]) for _ in range(cnt)])
        f.write(s + '\n')
    f.close()
    operate('Data.txt')

def operate(filename):
    
    # Поиск строки с самым большим нечётным числом
    
    f = open(filename, 'r')
    line, number = 0, 0
    maxline=f.readline() #   Строка с максимальным нечётным числом
    maxnum=0             #   Максимальное нечётное
    linenum=0
    maxlinenum=1
    for s in f:
        linenum+=1
        curnum=0    #    Текущее число
        wordlen=0   #    Длина текущего слова/числа
        for c in s: #    прогон по символам
            wordlen+=1
            if (ord(c)>=ord('0') and ord(c)<=ord('9')):
                curnum=curnum*10+(ord(c)-48)
            else:
                if ((curnum%2==1) and (curnum>=maxnum)):
                    maxnum=curnum
                    maxline=s
                    maxlinenum=linenum
                wordlen=0
                curnum=0
        if ((curnum%2==1) and (curnum>=maxnum)):
            maxnum=curnum
            maxlinenum=linenum
            maxline=s
    f.close()
    
    # Определение самой длинной возрастающей последовательности из букв
    
    maxword=''
    maxlenword=0
    curword=''
    curlenword=0      ## длина текущей последовательности
    curdesc=False     ## является ли последовательность убывающей
    for c in maxline: #     прогон по символам
        if (ord(c)>=ord('0') and ord(c)<=ord('9')):
            if curlenword>0 and curlenword>=maxlenword and not(curdesc):
                maxword=curword
                maxlenword=curlenword
            curword=''
            curlenword=0 
            curdesc=False

        else:
            if (not(curdesc)):
                if (curlenword==0): 
                    curlenword+=1
                    curword+=c
                else:
                    if (ord(c)>ord(curword[-1])): # следующий символ больше последнего в строке?
                        curlenword+=1
                        curword+=c
                    else:
                        curword=''
                        curlenword=0
                        curdesc=True
    if (curlenword>0 and curlenword>=maxlenword and not(curdesc)):
        maxword=curword
        maxlenword=curlenword

    # Заполнение файла Result.txt
    
    f = open('Result.txt', 'w')
    if maxlenword == 0:
        f.write('В файле нет возрастающих последовательностей\n')
    else:
        
    # Вывод результатов в файл
    
        f.write('Искомая строка - {}.\n'.format(linenum))
        f.write('Самое большое нечётное число - {}.\n'.format(maxnum))
        f.write('Самая большая возрастающая буквенная последовательность - {}.\n'.format(maxword))
    f.close()

#main
n = int(input('Введите 1 для генерации случайного файла. Введите иное число для обработки файла Test.txt'))
if n==1:
    randfill()
else: 
    filename='Test.txt'
    operate(filename)
````

## Задача B
 
Напишите программу, находящую наибольшие общие делители для всех переданных в стандартный поток последовательностей чисел.

Формат ввода
Вводятся строки чисел, разделённых пробелами.

Формат вывода
Последовательность чисел — наибольшие общие делители всех последовательностей.

## Листинг задачи B

````python
import sys


def gcd(a, b):
    if b == 0:
        return a
    else:
        return gcd(b, a % b)


for line in sys.stdin:
    lst = list(map(int, line.split()))
    if len(lst) > 0:
        result = lst[0]
        for i in range(1, len(lst)):
            result = gcd(result, lst[i])
        print(result)
````

````python
import numpy as np

import numpy as np


def snake(width, height, direction='H'):
    if direction in ('H', 'V'):
        if direction == 'V':
            width, height = height, width
        matr = [[x for x in range(y * width * (y % 2 == 0) + ((y - 1) * width + 1) * (y % 2),
                                  (y * width + 1) * (y % 2) + ((y - 1) * width) * (y % 2 == 0), (-1) ** (y - 1))] for y
                in range(1, height + 1)]

        if direction == 'V':
            matr = np.transpose(matr)
        matr = np.array(matr, dtype='int16')
        return matr


print(snake(3, 2, direction='V'))
````

## Листинг задачи J

````python
import numpy as np


def stairs(vector):
    matr = []
    newvec = vector
    for i in range(1, len(vector) + 1, 1):
        if i == 1:
            matr.append(vector)
        else:
            newvec = np.append(newvec[len(vector) - 1], newvec)
            newvec = newvec[0:len(vector):]
            matr.append(newvec)
    matr = np.array(matr)
    return matr

print(stairs(np.arange(5)))
````

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import random

%matplotlib notebook

xmin, xmax, hx = -np.pi, np.pi, 0.01

fig = plt.figure(figsize=(3, 3))
ax = plt.axes(xlim=(xmin, xmax), ylim=(-1.1, 1.1))

graph, = plt.plot([], [], 'b*')

x = np.arange(xmin, xmax, hx)

def update(x):
    graph.set_data(x, np.sin(x*random.randrange(-10, 11, 1))/4)
    return graph,

animation = FuncAnimation(fig,
                         update,
                         frames=np.arange(xmin, xmax, hx),
                         interval=1,
                         blit=True,
                         repeat=True
                         )
plt.show()
````
